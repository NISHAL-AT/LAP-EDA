```python
#SCATTER- REPRESENTING OUTLIER IN GRAPH USING SD
import numpy as np
z = np.abs((df['Price'] - df['Price'].mean()) / df['Price'].std())

outliers = df[z > 3]

plt.scatter(df['ram_memory'], df['Price'], color='grey', edgecolor='black', s=50)
plt.scatter(outliers['ram_memory'], outliers['Price'], color='black',edgecolor='grey', s=100, label='Outliers')
plt.xlabel('RAM (GB)')
plt.ylabel('Price')
plt.title('Outliers in Price vs RAM')
plt.legend()

plt.savefig('images/outliers_representation.png', dpi=300, bbox_inches='tight')  
plt.show()
plt.close() 
```

```python
#OUTLIERS- BASED ON PRICE(IQR)
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = df[(df['Price'] < lower_bound) | (df['Price'] > upper_bound)]
print("Outliers based on IQR:\n", outliers_iqr[['Price']])
```

```python
#OUTLIERS- Z TEST(BASED ON PRICE)
from scipy import stats
import numpy as np

z_scores = np.abs(stats.zscore(df['Price']))
outliers_z = df[z_scores > 3]

print("Outliers based on Z-score:\n", outliers_z[['Price']])
```

```python
#FEATURE EXTRACTION
# Total storage=primary + secondary
df['total_storage'] = df['primary_storage_capacity'] + df['secondary_storage_capacity']

# Total pixels=width x height
df['total_pixels'] = df['resolution_width'] * df['resolution_height']

# Price per GB RAM
df['price_per_ram'] = df['Price'] / df['ram_memory']

df[['primary_storage_capacity','secondary_storage_capacity','total_storage',
    'resolution_width','resolution_height','total_pixels','Price','ram_memory','price_per_ram']].head()
```

```python
#FEATURE SELECTION
numeric_cols = df.select_dtypes(include='number').columns

corr_matrix = df[numeric_cols].corr()

corr_with_price = corr_matrix['Price']

selected_features = corr_with_price[abs(corr_with_price) > 0.3].index.tolist()

selected_features.remove('Price')

print("Selected numeric features based on correlation with Price:")
print(selected_features)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew

#BEFORE
print("Original Skewness:", skew(df['Price']))
sns.histplot(df['Price'], kde=True)
plt.title("Original Price Distribution")
plt.show()

#AFTER TRANSFORMATION(LOG)
df['Price_log'] = np.log(df['Price'])
print("After Log Transform:", skew(df['Price_log']))
sns.histplot(df['Price_log'], kde=True)
plt.title("Price Distribution (After Log Transform)")
plt.show()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import skew

#BEFORE
print("Original Skewness:", skew(df['Rating']))
sns.histplot(df['Rating'], kde=True)
plt.title("Original Rating Distribution")
plt.show()

#AFTER TRANSFORMATION(SQUARE)
df['Rating_Square'] = df['Rating']**2
print("After Square Transform:", skew(df['Rating_Square']))
sns.histplot(df['Rating_Square'], kde=True)
plt.title("Rating Distribution (After Square Transform)")
plt.show()

#DUE TO MILD LEFT SKEW, TRANSFORMATION RESULTS IN MORE SKEWNESS, IN THAT CASE, TRANSFORMATION CAN BE AVOIDED AND THAT SKEWNESS IS NEGLIGIBLE AND CONSIDERED AS NORMAL DISTIRBUTION
```
